```python
import numpy as np
conv_kernel = np.array([[-1, 0, 1],
                        [-1, 0, 1],
                        [-1, 0, 1]])

data_matrix = np.array([
    [2, 2, 2, 2, 2, 1, 1, 1],
    [2, 2, 2, 2, 2, 1, 1, 1],
    [2, 2, 2, 2, 2, 1, 1, 1],
    [2, 2, 2, 2, 2, 1, 1, 1],
    [2, 2, 2, 9, 9, 9, 9, 9],
    [2, 2, 2, 9, 9, 9, 9, 9],
    [2, 2, 2, 9, 9, 9, 9, 9],
    [2, 2, 2, 9, 9, 9, 9, 9]
])
```

```python

def cov2(data_matrix, conv_kernel, stride, bias, padding, padding_h=1):
    if padding:
        img2d = np.pad(data_matrix,padding_h, mode='constant',
                     constant_values=0)
    else:
        img2d = data_matrix
    inw, inh = img2d.shape
    w, h = conv_kernel.shape
    outwidth = (inw - w) // stride + 1
    outheight = (inh - h) // stride + 1
    arrayy = np.zeros(shape=(outwidth, outheight))
    for i in range(outheight):
        for j in range(outwidth):
            s = 0
            for k in range(w):
                for l in range(h):
                    s += img2d[k + i * stride][l + j * stride] * conv_kernel[k][l]
            arrayy[i][j] = s + bias
    return arrayy


def maxpooling(data_matrix, pool_kernel, stride, padding, padding_h=1):
    if padding:
        img2d = np.pad(data_matrix,padding_h, mode='constant',
                     constant_values=0)
    else:
        img2d = data_matrix
    inw, inh = img2d.shape
    w, h = pool_kernel, pool_kernel
    outwidth = (inw - w) // stride + 1
    outheight = (inh - h) // stride + 1
    arrayy_max = np.zeros(shape=(outwidth, outheight))
    for i in range(outheight):
        for j in range(outwidth):
            s = []
            for k in range(w):
                for l in range(h):
                    small_s = img2d[k + i * stride][l + j * stride]
                    s.append(small_s)
            arrayy_max[i][j] = np.array(s).max()
    return arrayy_max


def avgpooling(data_matrix, pool_kernel, stride, padding, padding_h=1):
    if padding:
        img2d = np.pad(data_matrix,padding_h, mode='constant',
                     constant_values=0)
    else:
        img2d = data_matrix
    inw, inh = img2d.shape
    w, h = pool_kernel, pool_kernel
    outwidth = (inw - w) // stride + 1
    outheight = (inh - h) // stride + 1
    arrayy_avg = np.zeros(shape=(outwidth, outheight))
    for i in range(outheight):
        for j in range(outwidth):
            s = []
            for k in range(w):
                for l in range(h):
                    small_s = img2d[k + i * stride][l + j * stride]
                    s.append(small_s)
            arrayy_avg[i][j] = round(np.array(s).mean(), 4)
    return arrayy_avg
```

```python
# answer for question 3
conv_value = cov2(data_matrix=data_matrix, conv_kernel=conv_kernel.T, stride=1, bias=0.5,
                    padding='True', padding_h=1)
print(conv_value)  
```

```python
# answer for question 4
conv_value_stride2 = cov2(data_matrix=data_matrix, conv_kernel=conv_kernel.T, stride=2, bias=0.5,
                        padding='padding', padding_h=1)
print(conv_value_stride2)  
```

```python
# answer for question 5(a)
max_pool_ans = maxpooling(data_matrix=conv_value, pool_kernel=3, stride=1, padding='padding', padding_h=1)
print(max_pool_ans)  
```
